# Defind a function with If-Statement

```python
# Products within different ranges can be given different flags, which are stored within a new column
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df_ord_products_merged['price_range'] = df_ord_products_merged.apply(price_label, axis=1)
```

```python
df_ord_products_merged['price_range'].value_counts(dropna = False)
```

```python
df_ord_products_merged['price_range'].max
```

```python
# When too many users make Instacart orders at the same time, the app freezes. 
# The senior technical officer at Instacart wants you to identify the busiest hours of the day. 
# Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” 
# Create a new column containing these labels called “busiest_period_of_day.”
```

```python
df_orders_products_merged.head()
```

```python
# Creating a dataframe with the count of values for the column 'order_hour_of_day'
Periods_of_time = df_orders_products_merged['order_hour_of_day'].value_counts().rename_axis('Hours').reset_index(name='frequency_of_hours')
```

```python
Periods_of_time 
```

```python
# In other to be able to group the data into Most orders, Average orders and Fewest orders, I split the data using percentile as criteria.
# Creating a list showing the percentiles for the value of frequency_of_hours counts of the column 'order_hour_of_day' rewritten as Hours.
```

```python
Percentile = list(df_orders_products_merged['order_hour_of_day'].value_counts().quantile([0.25, 0.5, 0.75]))
```

```python
Percentile
```

```python
# Fewest orders list
fewest_ord = list(Periods_of_time.loc[Periods_of_time ['frequency_of_hours'] <= Percentile[0]]['Hours'])
```

```python
# Average orders list
Average_ord = list(Periods_of_time [(Periods_of_time ['frequency_of_hours'] > Percentile[0]) & (Periods_of_time ['frequency_of_hours'] < Percentile[2])]['Hours'])
```

```python
# Most orders list
Most_ord = list(Periods_of_time.loc[Periods_of_time['frequency_of_hours'] >= Percentile[2]]['Hours'])
```

```python
Most_ord
```

```python
index_F = df_orders_products_merged['order_hour_of_day'].isin(fewest_ord)
```

```python
index_avg = df_orders_products_merged['order_hour_of_day'].isin(Average_ord)
```

```python
index_Most = df_orders_products_merged['order_hour_of_day'].isin(Most_ord)
```

```python
 df_orders_products_merged.loc[index_F,'Busiest_period_of_day'] = 'Fewest orders'
```

```python
df_orders_products_merged.loc[index_avg,'Busiest_period_of_day'] = 'Average orders'
```

```python
df_orders_products_merged.loc[index_Most,'Busiest_period_of_day'] = 'Most orders'
```
